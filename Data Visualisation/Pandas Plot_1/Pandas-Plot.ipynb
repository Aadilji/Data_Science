```python
np.random.seed(20) 
df = pd.DataFrame(np.random.rand(1000,1), index=pd.date_range('2021-07-18' , periods=1000), columns=['data'])
```

```python
df.plot() 

# x-axis = date
## y-axis = values
```

```python
# cumsum = commulative data
df.data.cumsum().plot() 
```

```python
df.data.cumsum()
```

```python
# new column 
df['data1'] = df.data.cumsum()
```

```python
# "apply" operation with a function to store cube into the column  
df['data2'] = df['data'].apply(lambda x : x *3)
```

### scatter plot

```python
# relation between 2 colummns, so we use the scatter plot
df1.plot.scatter(x = 'sepal_width', y = 'petal_width')
```

```python
df1.plot.scatter(x = 'sepal_length', y = 'petal_width') 
```

```python
# draw the relationship between sepal_length and petal_width and graph olor should be based on petal_length
## if the length is more color of the graph is also more and if the length of the graph is less, color is also low
df1.plot.scatter(x = 'sepal_length', y = 'petal_width' , c = 'petal_length') 
```

```python
df1 
```

```python
# label and size 
df1.plot.scatter(x  ='sepal_length' , y  = 'sepal_width', label = 'Aadil' ,s = 180 ,c = 'petal_width')
```

#### hexabin - a graph simiar to scatter plot 

```python
#### hexabin - a graph simiar to scatter plot 
df1.plot.hexbin(x  ='sepal_length' , y  = 'sepal_width', label = 'sudh' ,gridsize = 10) 
                                                                                # yaha size ke jagah grid-size use hota hai
```
