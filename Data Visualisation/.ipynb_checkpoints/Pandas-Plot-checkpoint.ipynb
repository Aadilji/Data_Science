```python
import pandas as pd 
import numpy as np
import matplotlib
import seaborn as sns
%matplotlib inline       
```

#### %matplotlib ------------>> this line is for the people who have old version of jupyter notebook 

```python
np.random.seed(20) 
df = pd.DataFrame(np.random.rand(1000,1), index=pd.date_range('2021-07-18' , periods=1000), columns=['data'])
```

```python
df
```

```python
df.plot() 

# x-axis = date
## y-axis = values
```

```python
# cumsum = commulative data
df.data.cumsum().plot() 
```

```python
df.data.cumsum()
```

```python
# new column 
df['data1'] = df.data.cumsum()
```

```python
# "apply" operation with a function to store cube into the column  
df['data2'] = df['data'].apply(lambda x : x *3)
```

```python
df 
```

```python
# blue and green lines are mixing
df.plot() 
```

```python
df.plot(figsize =(10,10))  
```
