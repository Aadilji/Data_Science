#### Box-Plot:- It is used to find the outliers.

```python
#### Box-Plot
df.iplot(kind = 'box') 

                                    # 0  = Minimum
                                    # Q1 = 25% 
                                    # Q2 = 50% 
                                    # Q3 = 75% 
                                    # 1 = Maximum
```

Interqartile Range :- The range b/w the the q3 and q1 is called as IQR.<br>
max = Q3 + 1.5 IQR<br>
min = Q1 - 1.5 IQR

Remember:-<br> if i get a minimum outiers in the dataset it will be available below 0 in the graph.<br>
    if i get a maximum outiers in the dataset it will be available above 1 in the graph.
