```python
import seaborn as sns
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
sns.load_dataset('tips') 
```

```python
tips = sns.load_dataset('tips')
```

```python
# difference b/w tips vs total bill
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips)  
```

```python
# Grouping
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips , hue = 'smoker') 
### hue means color on the basis of, here color on the basis of smoker.
```

```python
# coloring on the basis of days..
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips , hue = 'day')  
```

##### style :- Grouping of Grouping

```python
# now on thusrday, give me the data , which will show on thursday who were smoker and who were not..
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips , hue = 'day', style= 'smoker') 
```

#### so blue with circles are smokers and blue with cross are non-smokers.

##### size 

```python
# size...
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips , hue = 'day', style= 'smoker', size= 'size') 
```

```python
#### seperate on the basis of size...
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips , size = 'size') 
```

```python
#### seperate on the basis of tip-size...
sns.relplot(x = 'tip' , y = 'total_bill' , data = tips , size = 'tip') 
```

#### catplot:- If we want to draw the relationship b/w categorical data and numerical data we uses catplot.

```python
# relationship b/w categorical data and numercial data ( day vs tip )
sns.catplot(x = 'day' , y = 'tip' , data =tips ) 
```

```python
sns.catplot(x = 'smoker' , y = 'tip' , data =tips ) 
```

##### here we can see that a smoker has given the maximum tip, as the outlier is the most in the side of smoker 

#### Box-Plot 

```python
sns.catplot(x = 'day' , y = 'tip' , kind = 'box' , data = tips)
```

```python
# who has given the more tip weather the male or female...
sns.catplot(x = 'day' , y = 'tip' , kind = 'box' , data = tips , hue = 'sex') 
```

##### now to see that this data-set is normally distributed or not .... ditplot

#### distplot 

```python
sns.distplot(x) 
```

```python
x = np.random.randn(10000)
sns.distplot(x) 
```

```python
x = np.random.randn(1000000)
sns.distplot(x) 
```

#### more the data, more the best normal distribution...

Here the graphs works on the principle of central limit Theorem...<br>
<br>
Central-Limit Theorem:-<br>
    if we are going to collect enough amount of samples from any distribution, so foresure, the collection<br>
    of those samples is going to showcase,this type of normal-distributions.

```python
# rand function does not give the normally distributed data...
x=np.random.rand(1000000) 
```

```python
sns.distplot(x)
```
