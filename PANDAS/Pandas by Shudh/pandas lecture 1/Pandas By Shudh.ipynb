# H5 file 

```python
form_file.head() 
```

```python
h5_data = pd.HDFStore('test.h5')      # creating an empty instance of h5 file
                                       # h5 file works as a key & value pair format.
                                   # we can't read h5 file directly as a text file, we can only access
                                  # it by programing language or an APIs
```

```python
h5_data['obj1'] = form_file      # allocating data into h5 file object.
```

```python
h5_data['obj1'] 
```

```python
h5_data.put('obj2',form_file,format='table')   #h5_data empty file mai,object create kra humne obj2 
                                            # phr us object mai dala form_file ka data in table format.
```

```python
h5_data.select('obj2',where=['index >=10 and index <=15'])  
                                # fir humne query run ki index(rows(10-15)) print krne ke liye..
```
