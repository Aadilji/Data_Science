#### Merge Operation 

```python
np.random.seed(23)
arr1 = np.random.randn(3,4)
np.random.seed(23)
arr2 = np.random.randn(3,4)
df11 = pd.DataFrame(arr1,columns=["A","B","C","D"],index = [0,1,2])
df22 = pd.DataFrame(arr2,columns=["A","B","C","D"],index = [3,4,5]) 
```

```python
df11 
```

```python
df22 
```

```python
pd.merge(df11,df22,on= "A")   # merge opeartion for only A columsns
                             # x columns df11 ka hai and y column df22 ka hai. 
```

```python
df1 = pd.DataFrame({'key': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df2 = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']}) 
```

```python
df1 
```

```python
df2 
```

```python
# Merging data on the basis of key, take left as left and df2 as right
pd.merge(df1,df2,on= 'key',how = 'left') 
```

```python
pd.merge(df1,df2,on= 'key',how = 'right') 
```

```python
# merge on basis of keys, but take both keys
pd.merge(df1,df2,on= 'key',how = 'outer')
```

```python
df1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
df2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
df1
```

```python
pd.merge(df1,df2, on= ['key1','key2'])
```

### Join Operation <difference b/w join and merge>
#### join is done on the basis of row while merge operation is done on the basis of column.

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2']) 

df2 = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3']) 
```

```python
df1 
```

```python
df2 
```

```python
df1.join(df2)  
```

```python
df1.join(df2, how='right')   
```

```python
df1.join(df2, how='outer') 
```
