# For Loops in File handeling 

```python
f = open("aadil.txt","r")      
content = f.read()

for line in content:
    print(line) 

f.close()    # character by character de rha hai which we dont want
```

```python
f = open("aadil.txt","r")      
for line in f:
    print(line) 

f.close()
```

```python
f = open("aadil.txt","r")      
for line in f:
    print(line, end="")         # end=blank

f.close()

#content = f.read()  ________ agar hum file pointer ko read karne ka bolenge phr iterate karenge toh 
# _____________________ ek baar toh iteration ho jaayega lekin uske baad pointer empty ho jaayega
# _____________________  and phr se iterate karne par woh blank(none) return karenga..
```

```python
f = open("aadil.txt","r")
f1 = open("aadil2.txt","w") 

for data_add in f:
    f1.write(data_add) 
    
f.close() # aadil.txt ka sara data copy ho kr aadil2.txt mai chala gya  hai
f1.close() 
 
```

```python
f = open('python.jpg','rb')

for i in f:
    print(i) 

f.close() 
```

```python
f = open('python.jpg','rb')         # read binary

f1 = open('mypython.jpg','wb')       # write binary

for i in f:
    f1.write(i)
    
f.close()
f1.close() 
```

# Readline() Function

```python
f = open("aadil.txt","rt")      
print(f.readline())

f.close()
```

```python
f = open("aadil.txt","rt")      
print(f.readline(),end="")
print(f.readline()) 

f.close()          # end="", it removes the /n after a line and replace it to a blank space.
```

# Readlines() function ...... to store all lines in a list. 

```python
f = open("aadil.txt","rt")      
print(f.readlines())
f.close()     # store all lines in a list,with new line characters and all other escape seq character.
```

# Writing in a File

```python
f = open("aadil.txt","w")      
f.write("aadil jeetega tu ek din")

f.close()    # agar aadil.txt pehle se hi bani hogi toh woh sab hata kr yeh line add kr dega
             # aur agr nhi hogi yeh file toh new file bana kar yeh likh dega
```

```python
f = open("aadil2.txt","w")      
f.write("aadil jeetega tu ek din")

f.close()    # agar hum yeh code pycharm me baar baar run krte rahenge toh yeh baar baar aadil.txt 
             # open karta raahega aur usme sab del kar ke baar baar yeh line likhte rahenga
```

# Appending a File 

```python
f = open("aadil.txt","a")       
f.write("aadil jeetega tu ek din \n") 

f.close()     # hum jeetni baar run karenge utni baar yeh append karta rahega
```

```python
f = open("aadil.txt","a")       
a = f.write("aadil jeetega tu ek din \n") 
print(a) 
f.close()     # yeh no. of character print karta hai jo humne file mai write kiye hai.
```

# Handle Read And Write Both 

```python
f = open("aadil.txt","r+")       
a = f.write("aadil jeetega tu ek din \n") 
print(f.read())
f.write("thank u") 
f.close() 
```

# tell() and seek() methods 

```python
# tell() method gives us the file pointer or handler position
f = open("aadil.txt")
print(f.tell())
print(f.readline())
print(f.tell()) 
print(f.readline())
print(f.tell()) 
f.close()
```

```python
# seek() method resets our pointer
f = open("aadil.txt")
print(f.readline())
f.seek(0)
print(f.readline())
f.close()                        # f.seek() ne reset kr diya pointer ko and phr code ne wapis 1st line read ki.
```

```python
f = open("aadil.txt")
print(f.readline())
f.seek(10)
print(f.readline())
f.close()  
```
