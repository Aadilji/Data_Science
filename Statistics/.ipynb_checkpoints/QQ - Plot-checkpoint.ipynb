```python
log_normal = np.random.lognormal(0,1,100)
print(log_normal)  
```

```python
sms.qqplot(log_normal , line= '45')  
plt.show() 
```

```python
sns.displot(log_normal, kind="kde") 
```

```python
sms.qqplot(np.log(log_normal), line = '45') 
plt.show() 
```
